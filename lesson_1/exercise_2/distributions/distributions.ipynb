```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('class_data.csv')
df
```

```python
df_subset = df[df["texts_yesterday"] < 100]
```

```python
df['texts_yesterday'].name = 'Amount'
texts_yesterday = df_subset.dropna()
plt.title('Texts yesterday')
sns.distplot(texts_yesterday["texts_yesterday"], kde=False) 
plt.show()
```

```python
mean_texts = df["texts_yesterday"].mean()
median_texts = df["texts_yesterday"].median()

print(f'The median amount of texts yesterday: {median_texts}')
print(f'The mean filesize of the texts_yesterday: {mean_texts} MB')
```

```python
sns.distplot(texts_yesterday['texts_yesterday'], kde=False) #I need to drop the NA's, otherwise Seaborn won't plot
plt.title('Texts yesterday')
plt.axvline(median_texts, 0, 100, color='black', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_texts, 0, 100, color='yellow', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
sns.distplot(texts_yesterday['texts_yesterday'])
plt.title('Texts yesterday')
plt.ylabel('Idontknowwhatthisis')
plt.show()
```

```python
sns.boxplot(df['texts_yesterday'])
plt.title('Texts yesterday')
plt.xlabel('Amount of texts')
plt.show()
```

```python
df_paid = df[df['texts_yesterday'] > 0] #Subsetting the dataframe to include only paid apps
sns.violinplot(df_paid['texts_yesterday'])
sns.swarmplot(df_paid['texts_yesterday'], color='purple') #default color doesn't stand out
plt.title('texts_yesterday')
plt.ylabel('texts_yesterday')
plt.xlabel('texts_yesterday')
plt.show()
```
