```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('googleplay_part1.csv', delimiter=';', encoding="utf-8")
df1.head()
```

```python
df2 = pd.read_csv('googleplay_part2.csv', encoding="utf-8")
df2.head(5)
```

```python
df = pd.merge(df1, df2)
df.head()
```

```python
df['waist'].value_counts()
```

```python
def under_20(x):
    if(x < 20): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(under_20)
```

```python
df["waist"].min()
```

```python
sns.distplot(df['waist'].dropna(), kde=False) #Selecting the rating column. I need to drop the NA's for the plot
plt.title('Waist size')
plt.xlabel('Waist size')
plt.show()
```
